```python
df = pd.read_csv("/content/gdrive/MyDrive/Hackathon_EFI/Clean Data/data_model.csv")
```

```python
df.set_index(["Ticker", "Date"], inplace=True)
```

```python
plt.subplots(1, 1, figsize=(12, 12))

ax = sns.heatmap(df.corr(), cmap="coolwarm", center=0, vmax=1, vmin=-1)
plt.show()
```

```python
df["log market cap"] = df["market cap"].apply(lambda x : np.log(1+x))
```

```python
INDUSTRY_MAIN = df.Industry.value_counts()[0:5].index
```

```python
sns.lmplot(
    data=(df[df.Industry.isin(INDUSTRY_MAIN)].reset_index()), 
    x="log market cap", 
    y="distance to default",
    col="Date",
    hue="Industry",
    col_wrap=3,
    ci=None, 
  )
```

```python
sns.lmplot(
    data=(df[df.Industry.isin(INDUSTRY_MAIN)].reset_index()), 
    x="log carbone emission", 
    y="distance to default",
    col="Date",
    hue="Industry",
    col_wrap=3,
    ci=None, 
  )
```

```python
sns.lmplot(
    data=(df[df.Industry.isin(INDUSTRY_MAIN)].reset_index()), 
    x="log carbone intensity", 
    y="distance to default",
    col="Date",
    hue="Industry",
    col_wrap=3,
    ci=None, 
  )
```

```python
sns.lmplot(
    data=(df[df.Industry == "Energy"].reset_index()), 
    x="log carbone emission", 
    y="carbone intensity",
    hue="distance to default",
    col="Date",
    palette="Reds",
    col_wrap=3,
    ci=None, 
  )
```

```python
df_graph = df.reset_index().set_index("Ticker")
df_graph = df_graph[df_graph.Industry == "Energy"]
for date in df_graph.Date.unique() : 
  filtre = df_graph.Date == date
  sns.scatterplot(
      data=df_graph[filtre],
      x="carbone emission", 
      y="carbone intensity",
      size="distance to default",
  )
  plt.show()

```
