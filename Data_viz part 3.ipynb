```python
filtre = df.Country.isin(df.Country.value_counts().index[0:6])

fig, ax = plt.subplots(1, 1, figsize=(15,8))

sns.barplot(
    data=df[filtre].groupby(["Industry", "Country"]).size().reset_index(),
    x="Industry",
    y=0,
    hue="Country",
    ax=ax)

ax.set_ylabel("Number of compagnies")
ax.set_title("Number of compagnies per country and per sector")

plt.xticks(rotation=45)
plt.show()
```

```python
#filtre = df["variable"].isin(df.variable.unique()[0:10])
fig, ax = plt.subplots(1, 1, figsize=(15,8))
ax.set_xlim(0,1.5)

sns.lineplot(
    data=df.groupby(["Industry", "Country"]).size().rese,
    x="carbone intensity",
    y="endog",
    hue="variable",
    ax=ax)

ax.set_xlabel("Carbone intensity")
ax.set_ylabel("risk")

plt.xticks(rotation=45)
plt.show()
```
