```python
path = "/content/gdrive/MyDrive/Hackathon_EFI/Clean Data/Sample 2007-2017/Carbone_Emission.xlsx"
df_carbone_emission = pd.read_excel(path)

path = "/content/gdrive/MyDrive/Hackathon_EFI/Clean Data/Sample 2007-2017/Carbone_Intensity.xlsx"
df_carbone_intensity = pd.read_excel(path)

path = "/content/gdrive/MyDrive/Hackathon_EFI/Clean Data/Sample 2007-2017/Distance to default.xlsx"
df_distance_to_default = pd.read_excel(path)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12,12))

ax = sns.heatmap(df_carbone_emission.isna())
plt.show()
```

```python
number = []

for t in range(0, len(df_carbone_emission)) : 
  filtre = df_carbone_emission.columns[(df_carbone_emission.loc[t:, :].isna().sum() == 0)]
  number.append(df_carbone_emission[filtre].shape[1])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7,4))

sns.lineplot(
    x=range(0, len(df_carbone_emission)),
    y=number,
    ax=ax
  )

ax.set_title("Number of compagnies according to the truncature")
plt.show()
```
